## Decision tree classification (Binary classification)

Problem Statement:Prediction of whether the games will be played or not? based on weather conditions

```python
import pandas as pd
```

```python
#Import the data from csv file into data frame
df=pd.read_csv("decision_tree_input.csv")
df.head(5)
```

```python
#As we know algorithm takes numerical values only,we need to convert our labels into numerical values
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
df['outlook1']=le.fit_transform(df.outlook)
df['humidity1']=le.fit_transform(df.humidity)
df['windy1']=le.fit_transform(df.windy)
df['will_play1']=le.fit_transform(df.will_play)
df
```

```python
df=df.drop(['outlook','humidity','windy','will_play'],axis=1)
df
```

# Split data into train and test dataset



```python
#From above we can see that our data is ready to input to model
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( df[['outlook1','humidity1','windy1']], df.will_play1, test_size=0.33)
```

```python
X_train
```

y_train

## Creating and training the model

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
#X_train=All attributes without taget variable
#y_train=target variable
```

```python
#now predict on test dataset
y_pred=model.predict(X_test)
```

## Measuring accuracy of model using confusion matrix

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
con_matrix=confusion_matrix(y_test,y_pred)
con_matrix
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

From above we can see that accuracy is 50%.Accuracy will increase with large volume of data
